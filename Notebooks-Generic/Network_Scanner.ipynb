```python
nm = nmap.PortScanner()
data = nm.scan(hosts="192.168.0.0/24", arguments="-sP")  # local network subnet
```

```python
hosts = []
for host_ip, host_info in data['scan'].items():
    host = {
        'ip': host_ip,
        'name': host_info['hostnames'][0]['name'],
        'type': host_info['hostnames'][0]['type'],
        'state': host_info['status']['state'],
        'reason': host_info['status']['reason'],
        'uphosts': data['nmap']['scanstats']['uphosts'],
        'ts': data['nmap']['scanstats']['timestr']
    }
    hosts.append(host)
hosts_df = pd.DataFrame(hosts)
hosts_df
```
